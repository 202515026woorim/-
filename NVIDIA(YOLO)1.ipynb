```python
import os

def download_video_with_ytdlp(url, filename="video.mp4"):
    # Shorts → 일반 링크 변환
    if "shorts/" in url:
        url = url.split("?")[0]
        url = url.replace("shorts/", "watch?v=")

    output = f"./{filename}"
    command = f'yt-dlp -f "best[ext=mp4]" -o "{output}" "{url}"'
    print(f"🎯 다운로드 명령어: {command}")
    os.system(command)

    if os.path.exists(filename):
        print(f"✅ 다운로드 성공: {filename}")
        return filename
    else:
        print(f"❌ 다운로드 실패")
        return None

# 예시 링크
youtube_url = "https://youtube.com/shorts/mzzOuzJ3E4I?si=lWxJxy9BMUINMg17"
video_path = download_video_with_ytdlp(youtube_url)
```
